```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('malariadeathdata.csv')
deaths = df.iloc[:,[3,4,7,9,23,29]]
```

```python
#filter out no malaria death countries
death = []
for i in deaths['Value']:
    if i == 'No malaria' or i == '0':
        death.append(False)
    else:
        death.append(True)
death_cleaned = deaths[death]
```

```python
#grouping countries into each region
AMR_death = death_cleaned[death_cleaned['ParentLocationCode'] == 'AMR']
AFR_death = death_cleaned[death_cleaned['ParentLocationCode'] == 'AFR']
EMR_death = death_cleaned[death_cleaned['ParentLocationCode'] == 'EMR']
SEAR_death = death_cleaned[death_cleaned['ParentLocationCode'] == 'SEAR']
WPR_death = death_cleaned[death_cleaned['ParentLocationCode'] == 'WPR']
```

```python
#calculating total deaths per region each year
AMR_death_total = AMR_death.groupby('Period').agg(np.sum)
AMR_death_total.reset_index()

AFR_death_total = AFR_death.groupby('Period').agg(np.sum)
AFR_death_total.reset_index()

EMR_death_total = EMR_death.groupby('Period').agg(np.sum)
EMR_death_total.reset_index()

SEAR_death_total = SEAR_death.groupby('Period').agg(np.sum)
SEAR_death_total.reset_index()

WPR_death_total = WPR_death.groupby('Period').agg(np.sum)
WPR_death_total.reset_index()

Death_total = death_cleaned.groupby('Period').agg(np.sum)
```

```python
#plotting each region's total deaths per year
fig, ax = plt.subplots(3,2, figsize = (15,15))

#graph with all regions data
ax[0,0].plot(AMR_death_total, label = 'AMR')
ax[0,0].plot(AFR_death_total, label = 'AFR')
ax[0,0].plot(EMR_death_total, label = 'EMR')
ax[0,0].plot(SEAR_death_total, label = 'SEAR')
ax[0,0].plot(WPR_death_total, label = 'WPR')
ax[0,0].legend(loc='upper right')
ax[0,0].set_ylabel('Deaths')
ax[0,0].set_xlabel('Year')
ax[0,0].set_title('Malaria Deaths by Region, 2000-2020')
ax[0,0].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

ax[0,1].plot(AMR_death_total)
ax[0,1].set_ylabel('Deaths')
ax[0,1].set_xlabel('Year')
ax[0,1].set_title('America Malaria Death Count 2000-2020')
ax[0,1].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

ax[1,0].plot(AFR_death_total)
ax[1,0].set_ylabel('Deaths')
ax[1,0].set_xlabel('Year')
ax[1,0].set_title('Africa Malaria Death Count 2000-2020')
ax[1,0].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

ax[1,1].plot(EMR_death_total)
ax[1,1].set_ylabel('Deaths')
ax[1,1].set_xlabel('Year')
ax[1,1].set_title('Eastern Mediterranean Malaria Death Count 2000-2020')
ax[1,1].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

ax[2,0].plot(SEAR_death_total)
ax[2,0].set_ylabel('Deaths')
ax[2,0].set_xlabel('Year')
ax[2,0].set_title('SE Asia Malaria Death Count 2000-2020')
ax[2,0].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

ax[2,1].plot(WPR_death_total)
ax[2,1].set_ylabel('Deaths')
ax[2,1].set_xlabel('Year')
ax[2,1].set_title('Western Pacific Malaria Death Count 2000-2020')
ax[2,1].set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])

fig.tight_layout(pad=5.0)
```
