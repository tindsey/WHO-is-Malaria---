# Analyzing Malaria Cases by Region - Lindsey To

```python
#import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing Malaria Cases Dataset

```python
#create the dataframe from the csv file 
malaria_df = pd.read_csv('total number of malaria cases.csv', dtype = 'string')

#Rename the columns to make it easier to work with 
malaria_cases =  malaria_df[['ParentLocation','Location','Period','FactValueNumeric']]
malaria_cases.columns = ['Region','Country','Year','Number of Cases']

print(malaria_cases)
```

## Subsetting Dataset by Region

```python
##subset the data by region, and make the region as rows and the rest leave as columns 

americas_df = malaria_cases[malaria_cases['Region']== 'Americas']
#set america as the row
row_index = 'Region'
americas_df = americas_df.set_index(row_index)

europe_df = malaria_cases[malaria_cases['Region']== 'Europe']
#set europe as the row
row_index = 'Region'
europe_df = europe_df.set_index(row_index)

africa_df = malaria_cases[malaria_cases['Region']== 'Africa']
#set africa as the row
row_index = 'Region'
africa_df = africa_df.set_index(row_index)

easternmed_df = malaria_cases[malaria_cases['Region']== 'Eastern Mediterranean']
#set eastern mediterranean as the row 
row_index = 'Region'
easternmed_df = easternmed_df.set_index(row_index)

westernpacific_df = malaria_cases[malaria_cases['Region']== 'Western Pacific']
#set western pacific as the row 
row_index = 'Region'
westernpacific_df = westernpacific_df.set_index(row_index)


sea_df = malaria_cases[malaria_cases['Region']== 'South-East Asia']
#set southeast asia as the row 
row_index = 'Region'
sea_df = sea_df.set_index(row_index)

#example of the new subset 
americas_df.head()
```

 ## Array of cases for each region per year

```python
#Create a for loop that produces the number of Malaria cases for Americas per year
america_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in americas_df.iterrows(): ## ind = Americas, rows = each row of Americas
        if (int(row['Year']) == int(year)):
            america_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Europe per year    
europe_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in europe_df.iterrows(): ## ind = Europe, rows = each row of Europe
        if (int(row['Year']) == int(year)):
            europe_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Africa per year    
africa_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in africa_df.iterrows(): ## ind = Africa, rows = each row of Africa
        if (int(row['Year']) == int(year)):
            africa_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Eastern Mediterranean per year
easternmed_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in easternmed_df.iterrows(): ## ind = Eastern Med , rows = each row of Eastern Med 
        if (int(row['Year']) == int(year)):
            easternmed_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1

#Create a for loop that produces the number of Malaria cases for the Western Pacific per year
westernpacific_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in westernpacific_df.iterrows(): ## ind = Western Pacific, rows = each row of WP
        if (int(row['Year']) == int(year)):
            westernpacific_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Southeast Asia per year
sea_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in sea_df.iterrows(): ## ind = SouthEast Asia, rows = each row of SEA
        if (int(row['Year']) == int(year)):
            sea_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
```

## Subplotting Malaria cases by region

```python
#generate figure with subplots 
figure,axis = plt.subplots(3,2, figsize =(20,20))

# plot americas malaria cases 
axis[0,0].plot(years,america_count,color ='blue')
axis[0,0].set_title('Malaria Cases in the Americas from 2010-2020')
axis[0,0].set_xlabel('Year')
axis[0,0].set_ylabel('Number of Cases')
axis[0,0].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[0,0].ticklabel_format(useOffset=True, style='plain')

#plot europe malaria cases 
axis[0,1].plot(years,europe_count,color='red')
axis[0,1].set_title('Malaria Cases in Europe from 2010-2020')
axis[0,1].set_xlabel('Year')
axis[0,1].set_ylabel('Number of Cases')
axis[0,1].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[0,1].ticklabel_format(useOffset=True, style='plain')

#plot africa malaria cases 
axis[1,0].plot(years,africa_count, color = 'purple')
axis[1,0].set_title('Malaria Cases in Africa from 2010-2020')
axis[1,0].set_xlabel('Year')
axis[1,0].set_ylabel('Number of Cases')
axis[1,0].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[1,0].ticklabel_format(useOffset=True, style='plain')

#plot eastern Mediterranean cases 
axis[1,1].plot(years,easternmed_count,color = 'pink')
axis[1,1].set_title('Malaria Cases in the Eastern Mediterranean from 2010-2020')
axis[1,1].set_xlabel('Year')
axis[1,1].set_ylabel('Number of Cases')
axis[1,1].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[1,1].ticklabel_format(useOffset=True, style='plain')

#plot western pacific cases 
axis[2,0].plot(years,westernpacific_count, color = 'black')
axis[2,0].set_title('Malaria Cases in the Western Pacific from 2010-2020')
axis[2,0].set_xlabel('Year')
axis[2,0].set_ylabel('Number of Cases')
axis[2,0].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[2,0].ticklabel_format(useOffset=True, style='plain')

#plot southeast asia cases 
axis[2,1].plot(years,sea_count, color = 'green')
axis[2,1].set_title('Malaria Cases in Southeast Asia from 2010-2020')
axis[2,1].set_xlabel('Year')
axis[2,1].set_ylabel('Number of Cases')
axis[2,1].set_xticks([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
axis[2,1].ticklabel_format(useOffset=True, style='plain')

plt.show()
```

## Plot demonstrating all regions from 2010-2020

```python
# plot each of the regions by year 
plt.plot(years, america_count)
plt.plot(years, africa_count)
plt.plot(years, europe_count)
plt.plot(years, easternmed_count)
plt.plot(years, westernpacific_count)
plt.plot(years, sea_count)

# show each year for x-axis ticks
plt.ticklabel_format(useOffset=True, style='plain')
plt.xticks(np.arange(min(years), max(years)+1, 1.0))

# label axis and title 
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.title('Malaria Cases from 2010-2020')

# generate legend
plt.legend(['America','Africa','Europe','Eastern Mediterranean','Western Pacific','Southeast Asia'])

plt.show()
```
