```python
#import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing Malaria Cases Dataset

```python
#create the dataframe from the csv file 
malaria_df = pd.read_csv('total number of malaria cases.csv', dtype = 'string')

#Rename the columns to make it easier to work with 
malaria_cases =  malaria_df[['ParentLocation','Location','Period','FactValueNumeric']]
malaria_cases.columns = ['Region','Country','Year','Number of Cases']

print(malaria_cases)
```

## Subsetting Dataset by Region

```python
##subset the data by region, and make the region as rows and the rest leave as columns 

americas_df = malaria_cases[malaria_cases['Region']== 'Americas']
#set america as the row
row_index = 'Region'
americas_df = americas_df.set_index(row_index)

europe_df = malaria_cases[malaria_cases['Region']== 'Europe']
#set europe as the row
row_index = 'Region'
europe_df = europe_df.set_index(row_index)

africa_df = malaria_cases[malaria_cases['Region']== 'Africa']
#set africa as the row
row_index = 'Region'
africa_df = africa_df.set_index(row_index)

easternmed_df = malaria_cases[malaria_cases['Region']== 'Eastern Mediterranean']
#set eastern mediterranean as the row 
row_index = 'Region'
easternmed_df = easternmed_df.set_index(row_index)

westernpacific_df = malaria_cases[malaria_cases['Region']== 'Western Pacific']
#set western pacific as the row 
row_index = 'Region'
westernpacific_df = westernpacific_df.set_index(row_index)


sea_df = malaria_cases[malaria_cases['Region']== 'South-East Asia']
#set southeast asia as the row 
row_index = 'Region'
sea_df = sea_df.set_index(row_index)

#example of the new subset 
americas_df.head()
```

 ## Array of cases for each region per year

```python
#Create a for loop that produces the number of Malaria cases for Americas per year
america_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in americas_df.iterrows(): ## ind = Americas, rows = each row of Americas
        if (int(row['Year']) == int(year)):
            america_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Europe per year    
europe_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in europe_df.iterrows(): ## ind = Europe, rows = each row of Europe
        if (int(row['Year']) == int(year)):
            europe_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Africa per year    
africa_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in africa_df.iterrows(): ## ind = Africa, rows = each row of Africa
        if (int(row['Year']) == int(year)):
            africa_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Eastern Mediterranean per year
easternmed_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in easternmed_df.iterrows(): ## ind = Eastern Med , rows = each row of Eastern Med 
        if (int(row['Year']) == int(year)):
            easternmed_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1

#Create a for loop that produces the number of Malaria cases for the Western Pacific per year
westernpacific_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in westernpacific_df.iterrows(): ## ind = Western Pacific, rows = each row of WP
        if (int(row['Year']) == int(year)):
            westernpacific_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
#Create a for loop that produces the number of Malaria cases for Southeast Asia per year
sea_count = [0]*11
indexOfCount = 0

for year in range (2010,2021):
    for ind, row in sea_df.iterrows(): ## ind = SouthEast Asia, rows = each row of SEA
        if (int(row['Year']) == int(year)):
            sea_count[indexOfCount] += int(row['Number of Cases'])
    indexOfCount += 1
    
```

## Import Packages and Clean Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('malariadeathdata.csv') #import csv file
deaths = df.iloc[:,[3,4,7,9,23,29]]  #extract the csv columns we need

#example of what data looks like
deaths.head()
```

```python
#filter out countries with 'no-malaria' or 0 deaths
death = []
for i in deaths['Value']:
    if i == 'No malaria' or i == '0':
        death.append(False)
    else:
        death.append(True)
death_cleaned = deaths[death]

#example of what data looks like
death_cleaned.head()
```

```python
#check that all values are a float, and we filtered out all no malaria countries
#european countries get filtered out, as there are no reported cases, or a very negligible number, of malaria deaths 

for i in list(death_cleaned['FactValueNumeric']):
    assert type(i) == np.float, "Data includes countries with No Malaria"
```

## Prepare Data for Graphing
Group data by regions.

```python
#separate data into new dataframes, based off ParentLocation (region)
america_df = death_cleaned[death_cleaned['ParentLocation'] == 'Americas'] 
africa_df = death_cleaned[death_cleaned['ParentLocation'] == 'Africa']
eastern_mediterranean_df = death_cleaned[death_cleaned['ParentLocation'] == 'Eastern Mediterranean']
southeast_asia_df = death_cleaned[death_cleaned['ParentLocation'] == 'South-East Asia']
western_pacific_df = death_cleaned[death_cleaned['ParentLocation'] == 'Western Pacific']

#example of what each dataframe looks like
america_df.head()
```

```python
#calculate total deaths/year for each region, from 2010-2020

AMR_death_total = america_df.groupby('Period').agg(np.sum) #sum up all America countries' data for each year
AMR_2 = AMR_death_total.reset_index() #reset index 
AMR_3 = AMR_2[AMR_2['Period']>= 2010] #extract years 2010-2020
AMR_death_total_final = AMR_3.set_index('Period') #set'Period' column back to index

#example of finished data
print(AMR_death_total_final)
```

```python
#repeat for all regions
AFR_death_total = africa_df.groupby('Period').agg(np.sum).reset_index()
AFR_2 = AFR_death_total[AFR_death_total['Period']>= 2010]
AFR_death_total_final = AFR_2.set_index('Period')

EMR_death_total = eastern_mediterranean_df.groupby('Period').agg(np.sum).reset_index()
EMR_2 = EMR_death_total[EMR_death_total['Period']>= 2010]
EMR_death_total_final = EMR_2.set_index('Period')

SEAR_death_total = southeast_asia_df.groupby('Period').agg(np.sum).reset_index()
SEAR_2 = SEAR_death_total[SEAR_death_total['Period']>= 2010]
SEAR_death_total_final = SEAR_2.set_index('Period')

WPR_death_total = western_pacific_df.groupby('Period').agg(np.sum).reset_index()
WPR_2 = WPR_death_total[WPR_death_total['Period']>= 2010]
WPR_death_total_final = WPR_2.set_index('Period')
```
